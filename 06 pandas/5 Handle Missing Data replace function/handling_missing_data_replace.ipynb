```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./weather_data.csv')
df
```

## Use replace() method to replace values in dataframe

```python
new_df = df.replace(-99999, np.NaN)
new_df
```

```python
new_df = df.replace([-99999, -88888], np.NaN)
new_df
```

```python
# Replace value based on specific value
new_df = df.replace({
    'temperature': -99999,
    'windspeed': -99999,
    'event': "0"
}, np.NaN)
new_df
```

## Replace values using a dictionary

```python
new_df = df.replace({
    -9999: np.NaN,
    '0': 'Sunny'
})

new_df
```

## How "regex" (regular expression) works

```python
# regex is used to replace pattern in string with new value in dataframe columns
new_df = df.replace('[A-Za-z]', '', regex=True)
new_df
```

## Replace data with "regex" using a dictionary

```python
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]'
}, '', regex=True)
new_df
```

## Replace the list of values with another list of values 

```python
df = pd.DataFrame({
    'A': ['a', 'b', 'c', 'd', 'e'],
    'B': ['f', 'g', 'h', 'i', 'j'],
})
df
```

```python
dfsss = df.replace(['a', 'b', 'c', 'd', 'e'], [1,2,3,4,5])
dfsss
```
